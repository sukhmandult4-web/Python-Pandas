```python
#Create a new DataFrame called patients that contains only these columns: Name, Age, Medical Condition, and Billing Amount.
patients=df[["Name","Age","Medical Condition","Billing Amount"]]
```

```python
patients
```

```python
#Filter the dataset to show only patients where Medical Condition is "Diabetes" and Age is greater than 50.
df[(df["Medical Condition"]=="Diabetes" )& (df["Age"] > 50)]
```

```python
#This dataset has some issues hiding in it:

# Remove exact duplicate rows using drop_duplicates().
# Some rows have a negative Billing Amount, which doesn't make sense for a hospital bill. Remove any rows where Billing Amount is less than 0.
# Print the number of rows before and after, so you can see how many were removed.

df.duplicated()
```

```python

# Some rows have a negative Billing Amount, which doesn't make sense for a hospital bill. Remove any rows where Billing Amount is less than 0.
df=df[df["Billing Amount"]>=0]
```

```python
df
```

```python
# Print the number of rows before and after, so you can see how many were removed.
df.shape
```

```python
#Look closely at the Name column — the capitalization is inconsistent (e.g. "Bobby JacksOn", "LesLie TErRy"). Clean it so every name is properly title-cased (e.g. "Bobby Jackson")#
df['Name'].str.title()
```

```python
#Convert Date of Admission and Discharge Date into real datetime columns
df["Date of Admission"]=pd.to_datetime(df["Date of Admission"])
df["Discharge Date"]=pd.to_datetime(df["Discharge Date"])
```

```python
df.info()
```

```python
#Then create a new column called Length of Stay that stores the number of days between admission and discharge for each patient.
df["Length of Stay"]=df["Discharge Date"]-df["Date of Admission"]
```

```python
df['Length of Stay']
```

```python
#Write a function called get_age_group(age) that returns:

# "Child" if age is 12 or under
# "Teen" if age is between 13 and 19
# "Adult" if age is between 20 and 59
# "Senior" if age is 60 or above

def get_age_group(age):
  if age <=12 :
    return "Child"
  elif 13 <= age <= 19:
    return "Teen"
  elif 20 <= age <= 59:
    return "Adult"
  else:
    return "Senior"
```

```python
#Then use .apply() to create a new column Age Group in df using this function.
df["Age_Group"]=df["Age"].apply(get_age_group)
```

```python
df["Age_Group"]
```

```python
#Using .apply() with a lambda function, create a new column called High Bill that stores "Yes" if a patient's Billing Amount is above the average billing amount of the whole dataset, and "No" otherwise.


df["High_Bill"]= df["Billing Amount"].apply(lambda  x: "HIGH" if x > df['Billing Amount'].mean() else "LOW ")
```

```python
df["High_Bill"]
```

```python
#Sort the dataset by Billing Amount in descending order, and display the top 10 most expensive cases  just the Name, Medical Condition, and Billing Amount columns.
df.sort_values(by="Billing Amount" , ascending=False).head(10) [["Name","Medical Condition","Billing Amount"]]
```

```python
#Group the data by Medical Condition and find the average Billing Amount and the average Length of Stay (from Question 8) for each condition.
df.groupby("Medical Condition")[["Billing Amount","Length of Stay"]].mean()
```
